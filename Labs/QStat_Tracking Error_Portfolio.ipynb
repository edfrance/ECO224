```python
import quantstats as qs
qs.extend_pandas()
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
stock=qs.utils.download_returns('NVDA',period='5y')
```

```python
print(type(stock))
```

```python
print(stock)
```

```python
stock.sharpe()
```

```python
qs.plots.snapshot(stock, title='Rendimiento de NVDA', show=True)
```

```python
qs.reports.html(stock,title='Analisis de retornos de NVDA',output='NVDA_report.html')
```

```python
qs.reports.html(stock,'SPY',title='Analisis de retornos de NVDA vs S&P500',output='NVDA_vs_S&P500_report.html')
```
